**Dataset:** Spotify Tracks Dataset — 114k tracks, 1921–2023  
**Business question:** *What audio and contextual features predict whether a track becomes popular — and how has the sound of popular music changed over time?*

```python
from datasets import load_dataset
from pathlib import Path
import pandas as pd

RAW = Path('../data/raw')

print('Downloading Spotify dataset from Hugging Face...')
ds = load_dataset('maharshipandya/spotify-tracks-dataset', split='train')
df = ds.to_pandas()

df.to_csv(RAW / 'spotify_tracks.csv', index=False)
print(f'Saved: {len(df):,} tracks')
print(df.dtypes)
```

```python
print('=== Dataset Overview ===')
print(f'Tracks:  {len(df):,}')
print(f'Columns: {len(df.columns)}')
print(f'Genres:  {df["track_genre"].nunique()}')
print(f'Artists: {df["artists"].nunique():,}')
print(f'\nNull counts:')
print(df.isnull().sum()[df.isnull().sum() > 0])
print('\nSample:')
df.head(3)
```
